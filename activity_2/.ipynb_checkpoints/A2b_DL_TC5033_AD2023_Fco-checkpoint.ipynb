```python
# Importing Libraries and Packages including PyTorch.

import numpy as np                              # Import numpy for linear algebra and array operations.
import torch                                    # Import torch library for deep learning framework.
import torch.nn as nn                           # Import torch.nn for neural network functions such as layers and loss functions.
import torch.nn.functional as F                 # Import pytorch functions for neural networks such as convolution, pooling, activation functions.
from torch.utils.data import DataLoader         # Import torch.utils.data for loading datasets for batching and shuffling.
from torch.utils.data import sampler            # Import torch.utils.data for sampling the datasets.
import torchvision.datasets as datasets         # Import torchvision.datasets for loading standard datasets such as MNIST, CIFAR10, etc.
import torchvision.transforms as T              # Import torchvision.transforms for image transformations such as cropping, flipping, etc.
import matplotlib.pyplot as plt                 # Import matplotlib.pyplot for plotting graphs.
```

```python
torch.cuda.is_available() #The torch.cuda.is_available() function is used to check if a GPU (Graphics Processing Unit) is available and can be used with PyTorch. It returns a boolean value.
```

```python
DATA_PATH = '/content/sample_data'                      # We define the path to store the data in google drive.
NUM_TRAIN = 50000                                       # We define the number of training samples.
NUM_VAL = 5000                                          # We define the number of validation samples.
NUM_TEST = 5000                                         # We define the number of test samples.
MINIBATCH_SIZE = 64                                     # We define the minibatch size for training and testing, could be 168 also.

# Transformations for the CIFAR10 dataset, we normalize the data with the mean and standard deviation of the dataset and convert it to tensor.
transform_cifar = T.Compose([
                T.ToTensor(),
                T.Normalize([0.491, 0.482, 0.447], [0.247, 0.243, 0.261])           # Normalizing the data with the mean and standard deviation of each channel of the dataset.
            ])                                                                      # The mean and standard deviation values allredy calculated and well known for this dataset.

# Train dataset.
cifar10_train = datasets.CIFAR10(DATA_PATH, train=True, download=True,              # We create the train dataset with the CIFAR10 dataset and apply the transformations defined above.
                             transform=transform_cifar)
train_loader = DataLoader(cifar10_train, batch_size=MINIBATCH_SIZE,                 # We create the train loader with the train dataset and define the minibatch size (WRAP).
                          sampler=sampler.SubsetRandomSampler(range(NUM_TRAIN)))    # We create a sampler to sample the data randomly from the train dataset.

# We repeat the same process for the validation and test datasets.

#Validation set.
cifar10_val = datasets.CIFAR10(DATA_PATH, train=False, download=True,
                           transform=transform_cifar)
val_loader = DataLoader(cifar10_val, batch_size=MINIBATCH_SIZE,
                        sampler=sampler.SubsetRandomSampler(range(NUM_VAL)))
#Test set.
cifar10_test = datasets.CIFAR10(DATA_PATH, train=False, download=True,
                            transform=transform_cifar)
test_loader = DataLoader(cifar10_test, batch_size=MINIBATCH_SIZE,
                        sampler=sampler.SubsetRandomSampler(range(NUM_VAL, len(cifar10_test))))
```

```python
# We a group of images from the test dataset and plot them.
def plot_cifar10_grid():
    classes = test_loader.dataset.classes
    total_samples = 8
    plt.figure(figsize=(15,15))
    for label, sample in enumerate(classes):
        class_idxs = np.flatnonzero(label == np.array(test_loader.dataset.targets))
        sample_idxs = np.random.choice(class_idxs, total_samples, replace = False)
        for i, idx in enumerate(sample_idxs):
            plt_idx = i*len(classes) + label + 1
            plt.subplot(total_samples, len(classes), plt_idx)
            plt.imshow(test_loader.dataset.data[idx])
            plt.axis('off')

            if i == 0: plt.title(sample)
    plt.show()

plot_cifar10_grid()
```

### Calcular accuracy


```python
# We iterate over the train loader to check the size of the minibatch and the number of minibatches.
for i, (x, y) in enumerate(train_loader):
    print(x, y)
```

```python
# We define a function to check the accuracy of the model.
def accuracy(model, loader):                                    # The data loader is passed as an argument to the function for the  minibatch.
    num_correct = 0
    num_total = 0
    model.eval()                                                # We set the model to evaluation mode.
    model = model.to(device=device)                             # We move the model to the device (GPU or CPU).
    with torch.no_grad():
        for xi, yi in loader:                                   # We iterate over the data loader to get the minibatch.
            xi = xi.to(device=device, dtype = torch.float32)
            yi = yi.to(device=device, dtype = torch.long)
            scores = model(xi)                                  # We run the model on the minibatch to get the scores.
            _, pred = scores.max(dim=1)                         # We get the predictions by taking the maximum score.
            num_correct += (pred == yi).sum()                   # We sum the number of correct predictions in the minibatch.
            num_total += pred.size(0)                           # We sum the total number of predictions in the minibatch.
        return float(num_correct)/num_total                     # We return the accuracy by dividing the number of correct predictions by the total number of predictions.
```

### Training Loop


```python
# We define a function to train the model.
def train(model, optimiser, epochs=100):                                # The model, optimiser and number of epochs are passed as arguments to the function.
    model = model.to(device=device)                                     # We move the model to the device (GPU or CPU).
    for epoch in range(epochs):                                         # We iterate over the number of epochs.
        for i, (xi, yi) in enumerate(train_loader):
            model.train()                                               # We set the model to training mode.
            xi = xi.to(device=device, dtype=torch.float32)              # We move the data to the device (GPU or CPU).
            yi = yi.to(device=device, dtype=torch.long)                 # We move the data to the device (GPU or CPU).
            scores = model(xi)                                          # We run the model on the minibatch to get the scores.
            cost = F.cross_entropy(input= scores, target=yi)            # We compute the loss function.
            optimiser.zero_grad()                                       # We set the gradients to zero.
            cost.backward()                                             # We compute the gradients of the loss function with respect to the parameters of the model.
            optimiser.step()                                            # We update the parameters of the model.
        acc = accuracy(model, val_loader)
        print(f'Epoch: {epoch}, costo: {cost.item()}, accuracy: {acc},')# We print the epoch, cost and accuracy for each epoch.
```

### Linear model

```python
# We define the hiddien layers, learning rate and number of epochs.
hidden1 = 256                   # We define the number of neurons in the first hidden layer.
hidden = 256                    # We define the number of neurons in the second hidden layer.
lr = 0.001                      # We define the learning rate.
epochs = 10                     # We define the number of epochs.
model1 = nn.Sequential(nn.Flatten(), # We flatten the input to the model.
                       nn.Linear(in_features=32*32*3, out_features=hidden1), nn.ReLU(), # We define the first hidden layer with 256 neurons and ReLU activation function.
                       nn.Linear(in_features=hidden1, out_features=hidden), nn.ReLU(),  # We define the second hidden layer with 256 neurons and ReLU activation function.
                       nn.Linear(in_features=hidden, out_features=10))                  # We define the output layer with 10 neurons.
optimiser = torch.optim.Adam(model1.parameters(), lr=lr)                                # We define the Adam optimiser with the learning rate defined above.
```

This defines a neural network model with two hidden layers, ReLU activations, and an output layer for classification. It also sets up the optimizer with the specified learning rate and prepares the model for training.

```python
# We train the model for a linear model.
train(model1, optimiser, epochs)
```

### Sequential CNN

**CNN Architecture:**

1. **Convolution and Pooling Stages:** 
    - The architecture consists of two stages of convolution and pooling. Each stage includes a convolutional layer followed by Batch Normalization and ReLU activation, followed by a MaxPooling layer to reduce the spatial resolution of features. This structure allows capturing local patterns at different levels of abstraction in the images. 
    - The inclusion of Batch Normalization after each convolutional layer stabilizes training, reduces the risk of overfitting, and enables the use of higher learning rates, thereby accelerating convergence and enhancing the model's generalization capability.

2. **Number of Channels:** In the first stage, 16 channels have been used in the convolutional layer, and in the second stage, 32 channels have been employed in both convolutional layers. Increasing the number of channels in deeper layers enables the extraction of more complex and abstract features.

3. **Flatten Layer:** After the convolutional and pooling stages, a Flatten layer has been included to convert the 2D features into a 1D vector, thus preparing the data for the fully connected layer.

4. **Fully Connected Layer:** At the end of the CNN, a fully connected layer has been added, performing the final classification. This layer takes the features extracted by the convolutional layers and maps them to the 10 output classes in CIFAR-10.

**Selected Hyperparameters:**

5. **Learning Rate (lr) and Epochs:** A learning rate of 0.001 has been configured, and a total of 10 epochs have been chosen for the initial training. These values are a reasonable choice to start the training process smoothly. However, it's important to note that these hyperparameters can be fine-tuned through cross-validation to optimize performance.

**Stability and Regularization:**
- Batch Normalization and the architecture itself provide stability during training and serve as a form of regularization, enhancing the model's ability to generalize to unseen data.


```python
# We set up a sequential model with 2 convolutional layers.
channel1 = 32              # We define the number of channels in the first convolutional layer.
channel2 = 64               # We define the number of channels in the second convolutional layer.
epochs = 10                 # We define the number of epochs.
lr = 0.0001                 # We define the learning rate.

# We define CNN model for two layers.
modelCNN1 = nn.Sequential(nn.Conv2d(in_channels=3, out_channels=channel1,
                                   kernel_size=3, padding=1),
                          nn.ReLU(),          # We define the first convolutional layer with 32 channels, kernel size of 3 and padding of 1.
                          nn.Conv2d(in_channels=channel1, out_channels=channel2,
                                    kernel_size= 3, padding=1),
                          nn.ReLU(),          # We define the second convolutional layer with 64 channels, kernel size of 3 and padding of 1.
                          nn.MaxPool2d(2, 2), # We define the max pooling layer with a kernel size of 2 and stride of 2, it divides to half the image.
                          nn.Flatten(),       # The output is flattened to convert it from a 3D tensor to a 1D vector.
                          nn.Linear(in_features=16*16*channel2, out_features=12) # At last we add a linear layer.
                                   )
optimiser = torch.optim.Adam(modelCNN1.parameters(), lr)   # We use the defualt optimizer, "Adam".
```

We define the first convolutional layer with 32 channels, 3x3 kernel size and ReLU activation function.



```python
# We train the model for a sequential model.
train(modelCNN1, optimiser, epochs)
```

```python
conv_k_3 = lambda channel1, channel2: nn.Conv2d(channel1, channel2, kernel_size=3, padding=1)
```

```python
class CNN_class4(nn.Module):
    def __init__(self, in_channel, channel1, channel2):
        super().__init__()
        self.conv1 = conv_k_3(in_channel, channel1)
        self.bn1 = nn.BatchNorm2d(channel1)

        self.conv2 = conv_k_3(channel1, channel2)
        self.bn2 = nn.BatchNorm2d(channel2)

        self.max_pool = nn.MaxPool2d(2,2)

    def forward(self, x):
        x = F.relu(self.bn2(self.conv2(F.relu(self.bn1(self.conv1(x))))))
        return self.max_pool(x)
```

```python
channel1 = 16
channel2 = 32
channel3 = 64
channel4 = 128
epochs = 10
lr = 0.001
modelCNN5 = nn.Sequential(CNN_class4(3,channel1, channel2),
                          CNN_class4(channel2, channel4, channel4),
                          nn.Flatten(),
                          nn.Linear(in_features=8*8*channel4, out_features=10))

optimiser = torch.optim.Adam(modelCNN5.parameters(), lr)
```

```python
train(modelCNN5, optimiser, epochs)
```

```python
accuracy(modelCNN5, test_loader)
```
